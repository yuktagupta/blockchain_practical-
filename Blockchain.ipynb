````python
#D. Create a blockchain, a genesis block and execute it.
import Crypto
import binascii
import datetime
import collections
from Crypto.PublicKey import RSA
from Crypto import Random
from Crypto.Hash import SHA
from Crypto.Signature import PKCS1_v1_5
import hashlib
from hashlib import sha256


class Client:
    def __init__(self):
        random = Crypto.Random.new().read
        # Creating new public key and private key
        self._private_key = RSA.generate(1024, random)
        self._public_key = self._private_key.publickey()
        self._signer = PKCS1_v1_5.new(self._private_key)

    @property
    def identity(self):
        return binascii.hexlify(self._public_key.exportKey(format='DER')).decode('ascii')


class Transaction:
    def __init__(self, sender, receiver, value):
        self.sender = sender
        self.receiver = receiver
        self.value = value
        self.time = datetime.datetime.now()

    def to_dict(self):
        if self.sender == "Genesis":
            identity = "Genesis"
        else:
            identity = self.sender.identity
        return collections.OrderedDict({
            'sender': identity,
            'receiver': self.receiver,
            'value': self.value,
            'time': self.time
        })

    def sign_transaction(self):
        private_key = self.sender._private_key
        signer = PKCS1_v1_5.new(private_key)
        h = SHA.new(str(self.to_dict()).encode('utf8'))
        return binascii.hexlify(signer.sign(h)).decode('ascii')


class Block:
    def __init__(self):
        self.verified_transactions = []
        self.previous_block_hash = ""
        self.Nonce = ""

    @staticmethod
    def display_transaction(transaction):
        tx_dict = transaction.to_dict()
        print("Sender: " + tx_dict['sender'])
        print('-----')
        print("Receiver: " + tx_dict['receiver'])
        print('-----')
        print("Value: " + str(tx_dict['value']))
        print('-----')
        print("Time: " + str(tx_dict['time']))
        print('-----')


# Instantiate client and create a transaction
Ninad = Client()
t0 = Transaction("Genesis", Ninad.identity, 500.0)

# Create genesis block
block0 = Block()
block0.previous_block_hash = None
block0.Nonce = None
block0.verified_transactions.append(t0)

# Blockchain list
TPCoins = []
TPCoins.append(block0)

# Display the blockchain
def dump_blockchain(blockchain):
    print("Number of blocks in chain: " + str(len(blockchain)))
    for x in range(len(blockchain)):
        block_temp = blockchain[x]
        print("Block #" + str(x))
        for transaction in block_temp.verified_transactions:
            Block.display_transaction(transaction)
        print('-' * 20)
        print("=" * 30)

# Dump the blockchain
dump_blockchain(TPCoins)
````

````python
#F. Add blocks to the miner and dump the blockchain.

import datetime
import hashlib

# Create a class with two functions
class Block:
    def __init__(self, data, previous_hash):
        self.timestamp = datetime.datetime.now(datetime.timezone.utc)
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.calc_hash()

    def calc_hash(self):
        sha = hashlib.sha256()
        hash_str = self.data.encode("utf-8")
        sha.update(hash_str)
        return sha.hexdigest()

# Instantiate the class
blockchain = [Block("First block", "0")]
blockchain.append(Block("Second block", blockchain[0].hash))
blockchain.append(Block("Third block", blockchain[1].hash))

# Dumping the blockchain
for block in blockchain:
    print(
        f"Timestamp: {block.timestamp}\n"
        f"Data: {block.data}\n"
        f"Previous Hash: {block.previous_hash}\n"
        f"Hash: {block.hash}\n"
        + "-"*60
    )
````

## ðŸ” 1. **Restricted Access**

This contract demonstrates how access to certain functions is limited to the owner using a custom `onlyOwner` modifier.

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract RestrictedAccess {
    address public owner;

    // Modifier to restrict access to owner
    modifier onlyOwner() {
        require(msg.sender == owner, "Access Denied: Not the owner");
        _;
    }

    // Set deployer as the owner
    constructor() {
        owner = msg.sender;
    }

    // Function only accessible by the owner
    function sensitiveFunction() public view onlyOwner returns (string memory) {
        return "Owner access granted!";
    }

    // Optional: Transfer ownership to another address
    function changeOwner(address newOwner) public onlyOwner {
        require(newOwner != address(0), "Invalid address");
        owner = newOwner;
    }
}

## ðŸ’¸ 2. **Withdrawal Pattern**

This contract lets users deposit and safely withdraw Ether using the pull-over-push pattern.

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.0;

contract WithdrawalPattern {
    mapping(address => uint) public balances;

    // Deposit Ether to contract
    function deposit() public payable {
        balances[msg.sender] += msg.value;
    }

    // Users withdraw their own funds
    function withdraw() public {
        uint amount = balances[msg.sender];
        require(amount > 0, "No balance to withdraw");

        // Reset balance before sending to prevent reentrancy
        balances[msg.sender] = 0;

        // Transfer Ether to sender
        payable(msg.sender).transfer(amount);
    }

    // View user's balance
    function getMyBalance() public view returns (uint) {
        return balances[msg.sender];
    }
}
